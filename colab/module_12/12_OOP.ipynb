## Object oriented programming

What is **object oriented programming (OOP)**?

### Classes

To create an object with methods, we use a **class** definition, which is a recipe defining the **attributes** and **methods** associated with the **class**. When we call the class, we create an **instance** of the **class**, also known as an **object**.

Here is an example of a **class** definition:

```python
class Circle: # The class keyword initiates the definition, followed by the class name (Circle)
    # a docstring (a string literal enclosed in triple quotes) is used to briefly explain the purpose of the class. 
    # This documentation can be accessed using Circle.__doc__ or help(Circle)
    """
    This is an example of a class called Circle
    """

    #These are variables defined directly within the class body, outside of any method. 
    # They are shared by all instances of the class.
    import numpy as np    
    pi = np.pi # pi is now an attribute of this class.
    
    # Initialize the class with the attribute r (no parentheses when called)
    # The __init__ method (initialization) is a special double-underscore method. 
    # It runs automatically every time a new instance of the class is created.
    # It sets up the initial state of the new object.
    # self is a convention used as the first parameter of all instance methods. 
    # It refers to the specific instance of the object being created or used.
    def __init__(self,r):
        self.r = r # define a variable, r
        
    # Define some methods (these have parentheses when called)
    # These are functions defined inside the class that perform operations using the object's data (self.r).
    # They must always accept self as their first parameter.
    # To call a method, you use dot notation on an instance (my_circle.area()). 
    # Python automatically passes the instance (my_circle) as the self argument.
    def area(self):
        """calculate the area"""
        return (1./2.)*self.pi*self.r**2
    
    def circumference(self):
        """calculate the circumference"""
        return 2.*self.pi*self.r
    
    def arclength(self,N):
        """calculate the arc length when dividing the circle into N equal parts"""
        return self.circumference()/N
```

Now we can create an **instance** of the Circle **class** called _cir_ with a radius of $r$.

```python
r = 3.5 # assign 3.5 to a variable r
cir = Circle(r) # create a class instance with radius of 3.5
```

We can use any of the attributes or methods of this class like this:

```python
print("The value of pi is: ",cir.pi) # no parentheses!
print("The radius of this circle is: ",cir.r) # no parentheses!
print("The area of a circle with radius = ",r,'is: ',cir.area()) # with parentheses!
print("The circumference of that circle is: ",cir.circumference()) # with parentheses!
print("One quarter arc length of that circle is: ",cir.arclength(4)) # with parentheses!
```

We can also save the Circle class in a module, just as we did for functions.

Then we can import it into other notebooks of scripts as desired.

```python
%%writefile Shapes.py
class Circle:
    """
    This is an example of a class called Circle
    """
    import numpy as np
    
    pi = np.pi # pi is now an attribute of this class.

    def __init__(self,r):
        self.r = r # define a variable, r
        
    def area(self):
        """calculate the area"""
        return (1./2.)*self.pi*self.r**2
    
    def circumference(self):
        """calculate the circumference"""
        return 2.*self.pi*self.r
    
    def arclength(self,N):
        """calculate the arc length when dividing the circle into N equal parts"""
        return self.circumference()/N
```

Now we can use it! Here is an example how: 

```python
import Shapes as S
cir = S.Circle(6.0)
print(cir.pi)
print(cir.r)
print(cir.circumference())
print(cir.arclength(4))
```

## Attributes and methods

You should have noticed that when we asked for **cir.pi** there were no parentheses, but **cir.area()**, **cir.circumference()** and **cir.arclength()** have parentheses. Why?

The answer is that __r__ and **pi** are **attributes**, while **area**, **circumference** and **arclength** are **methods**. Did you notice that the method definitions look like functions, but are inside the class definition. 

A **method** really is a function, but it is special in that it belongs to a **class** and works on the **instance** of the **class**. They can only be called by using the name of the **instance**, followed by a dot, followed by the **method** (with parentheses).


## More about classes

Classes are not the same as functions. Although the **Shapes** module can be imported just the same as any other module, to use it we first have to create a class **instance** (by doing **cir = Shapes.Circle(r)**). 

All _**methods** (parts that start with **def**), have an **argument** list. The first **argument** has to be a reference to the class instance itself, which is always **self**, followed by any variables you want to pass into the **method**.

The "**\_\_init\_\_**" method initializes the **instance** attributes. In the Circle class, the **\_\_init\_\_** method defined the **attribute** **r**, which gets passed in when the class is first called.

Asking for any **attribute** retrieves the current value of that **attribute**.  

But, attributes can be changed:

```python
print(cir.r)
cir.r = 7
print(cir.r)
```

To summarize:

The **methods** (**area**, **circumference** and **arclength**) are defined just like any function except the use of **self** as the first argument. This is required in all class method definitions. In the **Circle** class, no other variables are passed to **area** and **circumference** because the only one used is $r$, so the argument list consists of only **self**. Calling **area** and **circumference** requires no further arguments (the parentheses are empty) and the class returns the current values. In contrast, the **arclength** method has an input argument **n**, which must be specified when calling **arclength**.


```python
cir.area()
```

You can make a subclass (child) of the parent class which has all the attributes and methods of the parent, but may have a few attributes and methods of its own. You do this by setting up another class definition within a class.  

So, the bottom line about classes is that they are in the same category of things as variables, lists, dictionaries, etc. That is, they are "data containers" but with benefits. They hold data, and they also hold the methods to process those data.
