## Code blocks
We are about ready to start writing a "real" program. First we need to talk about the structure of a python program and the concept of 'code block'. 

Every programming language provides a way to group blocks of code together and execute them under certain conditions. Python uses indentation to define code blocks, which make the code more readable.

In Python, code block refers to **a collection of code that is in the same block or indent**. This is most commonly found in classes, functions, and loops.

Python uses tabs or four spaces for indentation. Jupyter notebooks try to indent for you, e.g., after a statement terminating in a colon. No matter which you choose, spaces or tabs, Be Consistent; otherwise Python will throw errors.

You can tell where code blocks are, because they are (1) on the same indent and (2) uninterrupted by other lines of code not on the same indent.

Python code blocks are useful because they tell you what pieces of code will run as a part of a loop, class, or function.

A common form for a code block starts with a _condition_ statement (if this is _True_) terminated by a colon (:). Most often, this condition statement would be followed by an indented code block that is executed, if the statement is _True_.

Let's look at a few examples:


```python
# the block below defines a function called code_block()
def code_block():
  # Everything in this function is part of the same code block
    x = 100
    return x
```

```python
# the block below is a for loop
for i in range(4):
  # Everything in this loop is part of the same code block
  print(i)
```

```python
def my_code():
    x = 5 # same block
    x = x + 4 # same block
    return x # same block

my_code() # Not on the same block of code as above
```

```python
pwd = 'a1b2c3'

# This is an if-else statement
# The lines print(‘Logging on …’) and print(‘Incorrect password.’) are two separate code blocks. 

if pwd == 'A1b2c3':
    print('Logging on ...')
else:
    print('Incorrect password.')

print('All done!')
```

Exceptions to the code indentation rules are:

- Statements can be continued on the next line with the continuation character $\backslash$ and the indentation of the following line is arbitrary. In Python 3, the backslash is not always required, but it is certainly recommended for clarity.  

- If a code block consists of a single statement, then that may be placed on the same line as the colon.  

- The command "break" breaks you out of the code block. Use with caution!

- There is a cheat that comes in handy when you are writing a complicated program and want to put in the code blocks but don't want them to DO anything yet: the command **pass** does nothing and can be used to stand in for a code block.


## Conditional statements

Conditions statements are statements like: "x is greater than y", which evaluate to either **True** or **False**.  

In Python we would use a **relational operator** __>__ for the "is greater than" test so the question "is x is greater than y?" is written as: **x>y**

Here is a list of frequently used relational operators:  

"==" means "equals"; A==B (Does A equal B?)

"!=" means "does not equal"; A!=B (Is A not equal to B?)

"<" means "less than"

"<=" means "less than or equal to"

">" means "greater than"

">=" means "greater than or equal to"

Conditions (like A==B) can be combined with **and** or **or** to make complex tests. For example **((A==B) and (C!=D))**. Both have to be true for the condition to evaluate as **True**.  
Alternatively for **((A==B) or (C!=D))**, only one has to be true for the condition to evaluate as **True**.

## "if" statements

In an **if** statement, the code-block is executed (**only once**) if the condition is _True_.

Here is a simple **if** statement:

```python
fruit = 'durian'
#fruit = input('enter a fruit name: ')

if (fruit == 'durian'):  
    print('yuck!')
else:
    print('yum!')
```

There are additional conditions that you can include in the **if** statements. In Python these are: **elif** (short for "else if") which means if the first **if** statement is _False_, but the statement following **elif** is _True_, then do something. Another is **else**, which means "otherwise". 

  Consider these examples:

```python
latitude = 32.7 # define a variable 'latitude'

if (latitude < 24): 
    print("Tropical region")
elif (latitude> 24 and latitude < 66):
    print("Temperate region")
else:
    print("Polar region")
```

```python
# airfare.py

age = int(input('How old are you? ')) # get user input and convert it to an integer

if age <= 2:
    print('free')
elif 2 < age < 13:
    print('child fare')
else:
    print('adult fare')
```

**in** is another Python _reserved word_ that checks if a particular value is "in" a **list** (or **tuple** or **set**):  

```python
michigan_colleges = ['MTU','NMU','UM'] # define a list of colleges in the state of Michigan
print(michigan_colleges)

if 'MTU' in michigan_colleges:  # if the string "MTU" is in the list, then 
    pass # don't do anything

if 'MSU' not in michigan_colleges:
    print ('MSU is missing from the list.')
    michigan_colleges.append('MSU')

elif 'NMU' in michigan_colleges: # if the last if statement is false, try this one
    print ('MSU and NMU both in list')
    
else: # if both statements are false, do this:
    print ("MSU in list but NMU isn't")
```

You can also use if-else in one single line. It is called an **ternary operator**, which defines a conditional expression. The most common usage is to make a terse, simple dependent assignment statement. In other words, it offers a one-line code to evaluate the first expression if the condition is true; otherwise, it considers the second expression.

Basic usage: **value_1 if condition else value_2**

You can also chain the the ternary operators by using multiple conditions

Let's look at some examples:

```python
x = 18

# It is a one line if-else statement that assigns a value to a variable based on a condition.
result = 'High' if x > 10 else 'Low'

print(result)
```

## "for" loops


The **for** loop allows you to step through a **list** (or **tuple**, among other things) sequentially. It assigns the first element to a variable name, say, $x$, perform some operation on it and then return to the top of the loop and assign the next element to the same variable name.  

For example, if we have a list $L = [1,2,3]$, this loop will assign each element in $L$ to the variable $x$, square it, and then print the result.  

```python
L = [1,2,3]

# step through the list assigning each element to the variable, x
for x in L: # x is assigned to each element
    print(x, x**2 )
```

As a second example of a **for** loop, we can use the **list** generated by the dictionary method **dict.keys()**:

```python
zipcodes = {'Houghton':'49931', 'Hancock':'49930', 'Chassell':'49916'}

# just use the list generator (we don't need the list() function)
for k in zipcodes.keys():
    print(k) # print k (key)
    print(zipcodes[k]) # print the value corresponding to k
```

What just happened? The **for** statement assigned each element in the **list** generated by the method **.keys( )** to the variable **k** and then printed each element out. 

The **for** statement loops through the entire **list**, one element at a time.

```python
# .items() return list of tuples. At each loop, the tuple is unpacked and assigned to k, v
for k, v in zipcodes.items(): 
    print(k)
    print(v)
```

## Nested loops
You can build a series of loops within loops (_nested loops_).    

Remember that everything within an indented code block is executed completely before moving on to the next code block.  


```python
zipcodes = {'Houghton':'49931', 'Hancock':'49930', 'Chassell':'49916'}

for k in zipcodes.keys():
    print(k)
    code = zipcodes[k]
    
    if code == '49931':
        print('The city with zip code 49931 is', k)
```

## List, Set, and Dictionary Comprehensions
Comprehensions in Python provide us with a short and concise way to construct new sequences (such as lists, set, dictionary etc.) using sequences which have been already defined. 

Python supports the following 4 types of comprehensions:

- List Comprehensions
- Dictionary Comprehensions
- Set Comprehensions
- Generator Comprehensions

### List comprehensions
List Comprehensions provide an elegant way to create new lists. The following is the basic structure of a list comprehension:

\[ _expr_ for val in collection if _condition_ \]

The filter condition can be omitted, leaving only the expression:

\[ _expr_ for val in collection ]

Let's look at a few examples. Given a list of strings, how do we filter out strings with length 2 or less and also convert them to uppercase?

```python
strings = ['a', 'as', 'bat', 'car', 'dove', 'python']

# create a new list containing the uppercase version of each string with length > 2
newstrings = [x.upper() for x in strings if len(x) > 2]
print(newstrings)
```

To find the strings with a length of at least 3 and containing the substring of 'a':

```python
strings = ['a', 'as', 'bat', 'car', 'dove', 'python']

# create a new list containing strings with length >= 3 and containing the letter 'a'
newstrings = [x for x in strings if len(x) >= 3 and 'a' in x]
print(newstrings)
```

Given a list of integers, how do we keep only the numbers that are divisible by 3 and multiply them by a factor of 5?

```python
numbers = range(0,40,4)
print(list(numbers))

# create a new list containing numbers that are multiples of 3, each multiplied by 5
newnumbers = [i*5 for i in numbers if i%3==0]
#newnumbers = [i*5 for i in numbers if not i%3] # this also works
print(newnumbers)
```

### Set and dict comprehensions

Set and dict comprehensions are a natural extension, producing sets and dicts in an idiomatically similar way instead of lists. A **dict comprehension** looks like this:

dict_comp = { _key-expr_ : _value-expr_ for value in collection if _condition_ }

Or (omitting the _if_ condition):

dict_comp = { _key-expr_ : _value-expr_ for value in collection }

A **set comprehension** looks like the equivalent list comprehension except with curly braces instead of square brackets:

set_comp = { _expr_ for value in collection if _condition_ }

Like list comprehensions, set and dict comprehensions are mostly conveniences, but they similarly can make code both easier to write and read.

Consider the list of strings from before. Suppose we wanted a set containing just the lengths of the strings contained in the collection; we could easily compute this using a set comprehension:


```python
strings = ['a', 'as', 'bat', 'car', 'dove', 'python']

# create a set containing the lengths of the strings (no duplicates)
unique_lengths = {len(x) for x in strings}
print(unique_lengths)
```

As a simple dict comprehension example, we could create a lookup map of these strings to their locations in the list:

```python
strings = ['a', 'as', 'bat', 'car', 'dove', 'python']

# create a dictionary mapping each string to its index in the list
loc_mapping = {val : index for index, val in enumerate(strings)}
print(loc_mapping)
```

Here the **enumerate** is a built-in function which returns a sequence of (index, value) tuples. This is very helpful for keeping the track of the index of the items when iterating over a sequence.

We can also use dict comprehension to invert the keys and values in a dictionary.

```python
zipcodes = {'Houghton':'49931', 'Hancock':'49930', 'Chassell':'49916'}

# create a new dictionary that inverts keys and values from the zipcodes dictionary
zipcodes_inv = { v:k for k, v in zipcodes.items() }
print(zipcodes_inv)
```

### Nested list comprehensions

Suppose we have a list of lists containing some English and Spanish names. You might have gotten these names from a couple of files and decided to organize them by language. Now, suppose we want to get a single list containing all names with two or more e’s in them.

We can use nested list comprehension for this:

```python
all_data = [['John', 'Emily', 'Michael', 'Mary', 'Steven'], ['Maria', 'Juan', 'Javier', 'Natalia', 'Pilar']]

# create a new list containing names with 2 or more 'e's
result = [ name for names in all_data for name in names if name.count('e') >= 2 ]
print(result)
```

The for parts of the list comprehension are arranged according to the order of nesting, and any filter condition is put at the end.
