# Course Syllabus

### Overview
This course introduces Python as a versatile and powerful programming language for geoscience applications. Widely adopted across the Earth and environmental science community, Python offers an open-source ecosystem with extensive libraries for data processing, analysis, visualization, and large-scale computing. The course covers core programming concepts and progresses to advanced techniques for handling and analyzing structured geoscience datasets. Students will engage in a hands-on, collaborative learning environment structured around a flipped classroom model, where in-class time is dedicated to coding exercises, problem solving, and peer learning.

### Course format
Each week has two classes: Tuesday is used for lecture and code demonstration while Thursday is dedicated for a self-guided, collaborative coding lab. Students are expected to review lecture notes before lectures on Tuesdays and work on coding assignments on Thursdays with help from the instructor. All lecture materials and assignments are provided as Jupyter Notebooks.

### Student learning outcome
By the end of this course, students will be able to:
* Apply core programming principles to build efficient, reproducible, and collaborative data workflows.
* Use Jupyter Notebooks to create interactive documents that integrate code, narrative text, and visualizations.
* Write clear, concise, and maintainable Python code for processing and visualizing common geoscience data formats.
* Perform basic statistical analysis and visualization using Python libraries.
