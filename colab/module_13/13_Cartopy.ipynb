## Introduction to maps

To create maps in Python, we will use the **cartopy** library. If you do not have **cartopy** installed, you can use conda package manager to install cartopy:

**conda install cartopy**

Here's what we need to make maps with **cartopy**:

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
from cartopy.mpl.ticker import LongitudeFormatter, LatitudeFormatter
from cartopy.mpl.gridliner import LONGITUDE_FORMATTER, LATITUDE_FORMATTER
from cartopy import feature as cfeature
import matplotlib.ticker as mticker
from matplotlib import cm
import warnings
warnings.filterwarnings('ignore')
```

There are many different types of maps used in the Earth Sciences. A map tries to represent something that is essentially 3D (a globe) onto a 2D medium (paper or a computer screen). So all maps, except those at the smallest scale, will increasingly distort the area as the scale increases.

When we choose a map projection, we seek the one that distorts the least for our purpose.  

Here we will use a few popular projections to make maps on both the global and local scale. Let's start with the global projections.

In **cartopy** maps are instances of the **Axes** figure class. They have many methods, for example outlining continents, national boundaries, and state boundaries and plotting geospatial data, such as sampling locations, earthquakes, and much else.

Before we can plot a set of coordinates, they must be transformed from latitudes and longitudes to map coordinates and then plotted like anything else in **matplotlib**.  

To make a map instance, we use the **projection** keyword in the **plt.axes( )** object (normally **ax**).

To set up the map, you need to know what type of projection you want (there are a growing number to choose from) and depending on the projection type, you need to know the map boundaries and other particulars.  

### Mercator Projection

We are starting with the Mercator projection. To do this, we set the projection keyword to '**ccrs.Mercator**'.

So we need: **central_longitude, min_latitude**, and **max_latitude**

We make the map instance with a call like:  

**ax = plt.axes(projection=ccrs.Mercator(central_longitude=180, min_latitude=-70, max_latitude=70, globe=None))**

We also have to put something on the map, so let's draw the coastlines by using the **coastlines()** method: 

**ax.coastlines()**

Here's our basic Mercator map of the whole world: 

```python
plt.figure(figsize=(6,3))
ax = plt.axes(projection=ccrs.Mercator(central_longitude=180, min_latitude=-70, max_latitude=70))
ax.coastlines()
```

We usually want lines of latitude (parallels) and longitude (meridians) on there as well, so we use the **gridlines( )** method with the keyword **crs=ccrs.PlateCarree( )** to make a **globe** object (usually called $g$ or $gl$)

Now we stick the latitudes and longitudes onto our globe object ($gl$ in the following example). We will need the **matplotlib.ticker** module (already imported as **mticker**) which has the useful attributes **mticker.FixedLocator( )** method which allows us to create a special object of latitudes or longitudes with the desired spacing using **np.arange()**. These can be attached to the figure object using figure attributes **g1.ylocator** and **g1.xlocator** for latitudes and longitudes respectively.  

The steps are:

1) Make the figure object ($ax$) with the desired projection using **plt.axes()**.

2) Put on the coastlines

3) Make the globe object ($g1$) using **ax.gridlines()**

4) Put on the latitude and longitudes by setting the **g1.ylocator** and **g1.xlocator** attributes to the desired **mticker.FixedLocator** object.  

Here we go: 

```python
# make the figure object
plt.figure(figsize=(6,3))
ax = plt.axes(projection=ccrs.Mercator(central_longitude=180, min_latitude=-70, max_latitude=70, globe=None))

# put on the coastlines
ax.coastlines()

# make the globe object
gl = ax.gridlines(crs=ccrs.PlateCarree(),linewidth=1.5,linestyle='dotted')

# put on the latitudes
gl.ylocator = mticker.FixedLocator(np.arange(-90,91,30))

# put on the longitudes
gl.xlocator = mticker.FixedLocator(np.arange(-180,181,30));
```

If you want the lat/long labels to show, you can modify your commands like this:

```python
plt.figure(figsize=(6,3))
ax = plt.axes(projection=ccrs.Mercator(central_longitude=180, min_latitude=-70, max_latitude=70, globe=None))
ax.coastlines()
g1 = ax.gridlines(crs=ccrs.PlateCarree(),linewidth=1.5,linestyle='dotted',draw_labels=True)
g1.ylocator = mticker.FixedLocator(np.arange(-90,91,30))
g1.xlocator = mticker.FixedLocator(np.arange(-180,181,60))

# set the formatters for the gridlines
g1.xformatter = LONGITUDE_FORMATTER
g1.yformatter = LATITUDE_FORMATTER
```

To turn labels off for the top, bottom, left or right sides, we can use attributes of the globe object like this: 


```python
plt.figure(figsize=(6,3))
ax = plt.axes(projection=ccrs.Mercator(central_longitude=180, min_latitude=-70, max_latitude=70, globe=None))
ax.coastlines()
ylabels = np.arange(-90,91,30)
xlabels = np.arange(-180.,181,60.)
g1 = ax.gridlines(xlocs=xlabels,ylocs=ylabels,crs=ccrs.PlateCarree(),linewidth=1.5,linestyle='dotted',draw_labels=True)
g1.xformatter = LONGITUDE_FORMATTER
g1.yformatter = LATITUDE_FORMATTER
g1.top_labels = False
```


Now maybe you want some color? You can set the color of the continents with:

**ax.add_feature(cfeature.LAND,color='orange')**

and color in the oceans with:

**ax.add_feature(cfeature.OCEAN,color='lightblue')**


```python
plt.figure(figsize=(6,3))
ax = plt.axes(projection=ccrs.Mercator(central_longitude=180, min_latitude=-70, max_latitude=70, globe=None))
ylabels=np.arange(-90,91,30)
xlabels=np.arange(-180.,181,60.)
g1 = ax.gridlines(xlocs=xlabels,ylocs=ylabels, crs=ccrs.PlateCarree(),linewidth=1.5,linestyle='dotted',draw_labels=True)
g1.xformatter = LONGITUDE_FORMATTER
g1.yformatter = LATITUDE_FORMATTER
g1.top_labels = False
ax.add_feature(cfeature.OCEAN,color='lightblue')
ax.add_feature(cfeature.LAND,color='orange')
ax.coastlines() # this has to come last, or it is buried by the ocean, land colors
```

The lakes are missing! You can change lake colors using the **cfeature.LAKES** attribute if you so desire. You can set the facecolor and edgecolor separately.  

If you want to put something ON the map, just take the latitude and longitude (or arrays of them) and convert them to map coordinates using the **transform=ccrs.Geodetic( )** keyword in the **plt.plot( )** command. These you plot just like any ordinary **matplotlib** plot, except you need a **ax.set_global( )** command after the plot, or you get a map near a given region/location.

Let's plot the position of San Diego as a big white star with the marker="\*", markersize=20, and color = 'white' arguments.  


```python
plt.figure(figsize=(6,3))
ax = plt.axes(projection=ccrs.Mercator(central_longitude=180.0, min_latitude=-70.0, max_latitude=70.0, globe=None))
ylabels=np.arange(-90,91,30)
xlabels=np.arange(-180.,181,60.)
g1 = ax.gridlines(xlocs=xlabels,ylocs=ylabels, crs=ccrs.PlateCarree(),linewidth=1.5,linestyle='dotted',draw_labels=True)
g1.xformatter = LONGITUDE_FORMATTER
g1.yformatter = LATITUDE_FORMATTER
gl.top_labels = False
ax.add_feature(cfeature.OCEAN,color='lightblue')
ax.add_feature(cfeature.LAND,color='orange')
ax.add_feature(cfeature.LAKES,facecolor='lightblue',edgecolor='black')
ax.coastlines()

San_lat = 33
San_lon = -117%360  # takes the west longitude and converts to 0=>360

ax.plot([San_lon],[San_lat],marker='*',color='white',markersize=20,transform=ccrs.Geodetic(),markeredgecolor='black')
ax.set_global() # we need this or weird things happen
ax.coastlines()
```

### Orthographic Projection

The Mercator is a nice classical map, but it distorts the map at high latitudes.

Another type of map projection is the orthographic projection which is much less distorted. The downside to this projection is that you cannot see the whole globe at once. To create an orthographic map, you initialize a map instance with the projection set to ccrs.Orthographic and a tuple of the central longitude and latitude. 

```python
plt.figure(figsize=(4,4))
ax = plt.axes(projection=ccrs.Orthographic(-75, 42))
ax.coastlines()
```

And finish the map as before.  

```python
plt.figure(figsize=(4,4))
ax = plt.axes(projection=ccrs.Orthographic(-75, 42))
San_lat = 33
San_lon = -117%360  # takes the west longitude and converts to 0=>360 using modular syntax
gl = ax.gridlines(crs=ccrs.PlateCarree(),color='black',linewidth=1,linestyle='dotted')
#gl.xlabels_top = False
gl.ylocator = mticker.FixedLocator(np.arange(-80,81,20))
gl.xlocator = mticker.FixedLocator(np.arange(-180,181,60));
gl.xformatter = LONGITUDE_FORMATTER
gl.yformatter = LATITUDE_FORMATTER
ax.add_feature(cfeature.OCEAN,color='lightblue')
ax.add_feature(cfeature.LAND,color='orange')
ax.add_feature(cfeature.LAKES,facecolor='lightblue',edgecolor='black')
ax.plot([San_lon],[San_lat],marker='*',color='white',markersize=20,transform=ccrs.Geodetic(),markeredgecolor='black')
ax.set_global()
ax.coastlines()
```

### Mollweide projection

One more global scale example of a map projection is the Mollweide projection. The Mollweide projection is always a global map centered on the equator, so all you need to specify is the central longitude (with the **central_longitude** keyword).

```python
plt.figure(figsize=(5,3))
ax = plt.axes(projection=ccrs.Mollweide(central_longitude=180))
San_lat = 33
San_lon = -117%360  # takes the west longitude and converts to 0=>360
gl = ax.gridlines(crs=ccrs.PlateCarree(),color='black',linewidth=1,linestyle='dotted')
gl.top_labels = False
gl.ylocator = mticker.FixedLocator(np.arange(-90,91,30))
gl.xlocator = mticker.FixedLocator(np.arange(0,400,30));
gl.xformatter = LONGITUDE_FORMATTER
gl.yformatter = LATITUDE_FORMATTER
ax.add_feature(cfeature.OCEAN,color='lightblue')
ax.add_feature(cfeature.LAND,color='orange')
ax.add_feature(cfeature.LAKES,facecolor='lightblue',edgecolor='black')
ax.plot([San_lon],[San_lat],marker='*',color='white',markersize=20,transform=ccrs.Geodetic(),markeredgecolor='black')
ax.set_global()
ax.coastlines()
```

### Lambert conformal 

The maps we've explored so far are good for global scale problems, for example plotting the locations of earthquakes around the globe, but not so great for more local problems, like a map of sampling sites.

For this we need a smaller scale map and the Lambert conformal conic projection is a popular choice. For this we must specify the central latitude and longitude and the map boundaries (with **ax.set_extent()**.

Here's what we know so far:

```python
plt.figure(figsize=(5,5))
proj = ccrs.LambertConformal(central_longitude=260, central_latitude=33)
ax = plt.axes(projection=proj)
ax.set_extent([-130, -70, 20, 52], crs=ccrs.PlateCarree())
ax.add_feature(cfeature.OCEAN,color='lightblue')
ax.add_feature(cfeature.LAND,color='orange')
ax.add_feature(cfeature.LAKES,facecolor='lightblue',edgecolor='black',linewidth=1)
ax.plot([San_lon],[San_lat],marker='*',color='white',markersize=20,transform=ccrs.Geodetic(),markeredgecolor='black')
gl = ax.gridlines(ylocs=np.arange(0,90,15.),xlocs=np.arange(-180.,180,15.),linewidth=1.5, linestyle="dotted")

ax.coastlines()
```

It would be nice to put on national and state boundaries. We can draw the national boundaries in a thick line like this:  

```python
proj = ccrs.LambertConformal(central_longitude=260, central_latitude=33)
fig = plt.figure(figsize=(5,5), frameon=True) # you need this frameon to be true
ax = plt.axes(projection=proj)

ax.set_extent([-130, -70, 20, 52], crs=ccrs.PlateCarree())
ax.add_feature(cfeature.OCEAN,color='lightblue')
ax.add_feature(cfeature.LAND,color='orange')
ax.add_feature(cfeature.LAKES,color='lightblue',linewidth=1)
ax.plot([San_lon],[San_lat],marker='*',color='white',markersize=20,transform=ccrs.Geodetic(),markeredgecolor='black')

xticks=list(range(-180,-30,15))
yticks=list(range(0,90,15))

# Enable labels in gridlines
gl = ax.gridlines(ylocs=yticks,xlocs=xticks,linewidth=1.5, linestyle="dotted",draw_labels=True)
gl.xformatter = LONGITUDE_FORMATTER
gl.yformatter = LATITUDE_FORMATTER
gl.xlabel_style = {'size':8, 'color': 'black'}
gl.ylabel_style = {'size':8, 'color': 'black'}

ax.coastlines();
ax.add_feature(cfeature.BORDERS,linestyle='-',linewidth=2)

# Add states and provinces lines
states_provinces = cfeature.NaturalEarthFeature(
        category='cultural',
        name='admin_1_states_provinces_lines',
        scale='50m',
        edgecolor='purple',
        facecolor='none',
        linestyle='dotted')
ax.add_feature(states_provinces)
```

## Gridding

Let's start with an example of plotting elevation on orthographic projection.

From the previous module, we know how to create an orthographic map:

```python
plt.figure(1,(4,4))
ax = plt.axes(projection=ccrs.Orthographic(-75, 42))
San_lat = 33
San_lon = -117%360  # takes the west longitude and converts to 0=>360
gl = ax.gridlines(crs=ccrs.PlateCarree(),color='black',linewidth=1,linestyle='dotted')
#gl.xlabels_top = False
gl.ylocator=mticker.FixedLocator(np.arange(-80,81,20))
gl.xlocator=mticker.FixedLocator(np.arange(-180,181,60))
gl.xformatter = LONGITUDE_FORMATTER
gl.yformatter = LATITUDE_FORMATTER
ax.set_global()
ax.coastlines()
```

Let's import the topographic data: 

```python
etopo = np.loadtxt('https://raw.githubusercontent.com/ixnix/GE5280PythonGeosciences/main/module_13/data/etopo20data.gz')
elons = np.loadtxt('https://raw.githubusercontent.com/ixnix/GE5280PythonGeosciences/main/module_13/data/etopo20lons.gz')
elats = np.loadtxt('https://raw.githubusercontent.com/ixnix/GE5280PythonGeosciences/main/module_13/data/etopo20lats.gz')
print(etopo.shape)
print(elons.shape)
print(elats.shape)
```

**elats** is a 1D array with 540 latitudinal bins. **elons** is a 1D array with 1081 longitudinal bins. And **etopo** is a 2D array with 540 rows and 1081 columns (540 x 1080).  

So **etopo** has an elevation for each lat/lon cell.  

In order to plot the elevation data onto a lat/lon grid, we have to first make a grid out of the 1D arrays of elats/ and elons.  

We use the **numpy** function **meshgrid** for this.  


Let's try out meshgrid on some smaller arrays first 

```python
x = np.arange(-3.2, 3.3, 0.1) 
y = np.arange(-3.2, 3.3, 0.1)
xx, yy = np.meshgrid(x, y)
print(x.shape)
print(y.shape)
print(xx.shape)
```

Now let's put some 'data' on this grid:

```python
z = np.sin(xx**2 + yy**2) / (xx**2 + yy**2)
print(z.shape)
```

## Colormesh

Now we want to make a plot of the data. There are many options to do this. One rather simple way would be to plot the x,y data points and use color to represent the value of the z. For this we have **plt.pcolormesh( )**.  

```python
plt.figure(figsize=(3,3))
plt.pcolormesh(x,y,z,cmap=cm.turbo)
plt.axis('equal')
```

We probably want a color bar too:  

```python
plt.figure(figsize=(3.7,3))
plt.pcolormesh(x,y,z,cmap=cm.turbo)
plt.axis('equal')
plt.colorbar()
```

And the plot looks a little 'pixelated' (rough). We can smooth it with the keyword **shading**.  

```python
plt.figure(figsize=(3.7,3))
plt.pcolormesh(x,y,z,cmap=cm.turbo,shading='gouraud')
plt.axis('equal')
plt.colorbar()
```

A related technique to plotting on a mesh is contouring. Contouring breaks the data down into regions separated by equal values (contours). 

## Contouring

Another way to visualize data in a 2D grid would be to make a contour plot of it. For this we use the **matplotlib.pyplot** function **contour** or **contourf**.  

```python
plt.figure(figsize=(3,3))
h = plt.contour(x,y,z,colors='r') 
plt.axis('equal')
# add contour values
plt.clabel(h, inline=True, fontsize=8)
```

```python
# To apply a colormap to z values
plt.figure(figsize=(3,3))
h = plt.contour(x,y,z,cmap=cm.turbo) 
plt.axis('equal')
```

If you want a continuous gradation of color, not the contour lines, you can use the function **contourf** instead:

```python
plt.figure(figsize=(3.7,3))
fig = plt.contourf(x,y,z,cmap=cm.turbo)
plt.axis('equal') 
plt.axis('off')
plt.colorbar()
```

Note the difference between the **colormesh** (continuous gradation) and **contourf** (constant color between contours).  

Back to our original problem - we wanted to plot the contoured elevation data onto an orthographic projection.

```python
plt.figure(figsize=(4,3))
ax = plt.axes(projection=ccrs.Orthographic(-75, 42))

# Add gridlines
gl = ax.gridlines(crs=ccrs.PlateCarree(),color='black',linewidth=1,linestyle='dotted')
gl.ylocator = mticker.FixedLocator(np.arange(-80,81,20))
gl.xlocator = mticker.FixedLocator(np.arange(-180,181,60))
gl.xformatter = LONGITUDE_FORMATTER
gl.yformatter = LATITUDE_FORMATTER

X, Y = np.meshgrid(elons,elats)
levels = np.arange(-10000,8000,500) # define contour intervals

m = ax.contourf(X, Y, etopo,levels,transform=ccrs.PlateCarree(),cmap=cm.turbo)
ax.set_global()

# Add a colorbar
cbar = plt.colorbar(m, ax=ax, orientation='vertical', shrink=0.9)
cbar.set_label('Elevation (m)')
```
