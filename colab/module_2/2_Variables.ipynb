## Variables

Computer programs need to pass numbers around, manipulate them, save them, and so on. This is done by assigning values to variables. There are many variable types in Python, including: 

- integer (a number without a decimal),
- floating point (a number with a decimal),
- string (numbers and/or letters enclosed in quotation marks),
- complex numbers,
- booleans (True or False, 1 or 0).  

Before you can use a variable, it must be defined. The following code block shows how to define variables in Python.

```python
# Remember that Python ignores everything on a line after a pound sign (#)
# This is how you write "comments"
# I encourage you to use comments where you feel you will have a hard time remembering the function of tricky lines
# Believe me, it happens quite often that I have a hard time in understanding my own codes from the past.

iNumber = 1  # an integer
fNumber = 1.0  # a floating point - notice the decimal point
sNumber = '1'  # a string - notice the quotation marks
snumber_another = "1"  # double quotes are also ok
comp = 1j  # a complex number with imaginary part 1
morecomplex = 3 + 1j  # the complex number 3+1i
bools = True  # A boolean variable (True/False or 1/0)
```

bools is of boolean variable type which has two possible values: True, False or alternatively 1 and 0. Booleans have many uses but in particular can be used to control the flow of the program as we shall learn later.

To see what value has been assigned to a variable, simply type: **print** and then the variable name in parentheses.

```python
print(comp)
print(bools)
```

## Variable naming rules
There are some guidelines and rules in choosing variable names in Python.

1) Variable names are composed of alphanumeric characters, including '-' and '\_'.

2) They are case sensitive: 'a' is not the same as 'A'.

3) There are some _reserved words_ in Python that you may not use as your variable names because they have pre-defined meanings (for example, _False_ or _True_)

    Here is a list of **reserved words**: 

    **and**, **assert**, **break**, **class**, **continue**, **def**, **del**, **elif**, **else**, **except**, **exec**, **finally**, **for**, **from**, **global**, **if**, **import**, **in**, **is**, **lambda**, **not**, **or**, **pass**, **print**, **raise**, **return**, **try**, **while**

    Do NOT use any of the following words either, which conflict with the names of commonly-used Python functions:

    **Data**, **Float**, **Int**, **Numeric**, **array**, **close**, **float**, **int**, **input**, **open**, **range**, **type**, **write**, **zeros**

    You should also avoid all the names defined in commonly-used Python code libraries like: 

    **acos**, **asin**, **atan**, **cos**, **e**, **exp**, **fabs**, **floor**, **log**, **log10**, **pi**, **sin**, **sqrt**, **tan**

4) There are few rules for variable names except for avoiding reserved words, but there are "best practices". In general, longer more descriptive words are better because they'll remind you what the variable stores. 

5) A best practice is to use expressive variable names:

[Make Your Code Easier to Read By Using Expressive Variable Names in Python](https://www.earthdatascience.org/courses/intro-to-earth-data-science/write-efficient-python-code/intro-to-clean-code/expressive-variable-names-make-code-easier-to-read/)


Here are some popular choices for variable names:  

1) Use these lower case options for **variables**, **module** and **package**:
- lowercase  
- lower_case_with_underscores  
- mixedCase 
- Modules should have short, all-lowercase names. Underscores can be used in the module name if it improves readability.

2) Use these upper case options for **constants**
- UPPERCASE  
- UPPER_CASE_WITH_UNDERSCORES
- Constants are usually defined on a module level and written in all capital letters with underscores separating words. e.g., MAX_OVERFLOW and TOTAL.

3) Use CapitalizedWords (or CapWords, StudlyCaps, or CamelCase -- so named because of the bumpy look of its letters) for **classes**
- CapitalizedWords
- Do NOT use Capitalized_Words_With_Underscores (ugly!)

4) Don't use this, it's confusing.
- Capitalized_Words_With_Underscores 

5) Some things to avoid:

- Don't use characters 'l' (lowercase letter el), 'O' (uppercase letter oh), or 'I' (uppercase letter eye) as single character variable names. These are easily confused in some fonts with '1' (one), '0' (zero), for example.
- Don't use non-standard symbols like $^{\circ}$ or $\sim$.  
- Be careful with special names that use leading or trailing underscores. These are treated differently by Python and you have to know what you are doing before you use them.  

Variables are names for objects within a particular namespace; the type information is stored in the object itself. Use **type** command to find out the variable type.

```python
print(type(fNumber))
print(type(sNumber))
print(type(bools))
```

Let's try using some operations on the above defined variables. 

```python
print(iNumber+iNumber)  # adding two integers produces an integer
print(iNumber+fNumber) # adding an integer and a float makes a float
print(sNumber+sNumber) # adding two strings concatenates the strings
```

## String operations
strings can be denoted with single, double or triple quotes: 

```python
string1 = 'spam'
string2 = "Sam's spam"
print(string1)
print(string2)
```

You can also use triple quotes: 

```python
print("""  
Hi there this is the second week
of the new Python course.
I am getting good at it.
""")
```



Strings can be added together: 



```python
mystring = 'michigan' + 'tech'
print(mystring)
```

Strings can be repeated using *

```python
mystring = 'michigan'
print(mystring * 3)
```

The built-in function len() returns the length of a string:

```python
print(len(mystring*3))
```

Strings can be _indexed_. 
- The first index is 0.
- Indices may also be negative numbers, to start counting from the right

```python
mystring = 'michigan'

print(mystring[0])
print(mystring[-1])
```

Strings can be _sliced_. Slicing allows you to obtain substring.

```python
mystring = 'michigan'

print(mystring[0:2])
print(mystring[0:2])  
print(mystring[2:5])  
```

Note how the start is always included, and the end always excluded. This makes sure that <mark>s[:i] + s[i:]</mark> is always equal to <mark>s</mark>:

Slice indices have useful defaults; an omitted first index defaults to zero, an omitted second index defaults to the size of the string being sliced.

```python
mystring = 'michigan'

print(mystring[:2])
print(mystring[2:])
print(mystring[-2:])
```

Python strings cannot be changed — they are _immutable_. Therefore, assigning to an indexed position in the string results in an error. That means, you can't do this: 

```python
mystring[0] = 'b'
```

What if the string contains single or double quotes?

```python
newstring = 'I\'m getting better at Python'
print(newstring)
```


To find more of the things you can and cannot do to strings, see: http://docs.python.org/tutorial/introduction.html#strings


## Let's play with some variables


```python
a = 2
print(a)
```

```python
b = 2
print(b)
```

```python
c = a + b
print(c)
```

You will recognize $a, b,$ and $c$ in the above session as _variables_ and $+$ as an _operation_. And these examples are pretty straight-forward math operations.  

But programming operations are not the same as arithmetic ones. For example,

```python
print('c = ', c)
c = c + 1
print('now c is: ', c)
```

The trick here is that the right hand side gets evaluated first, then assigned to the left hand side.  

And here is another funny looking statement, which is also perfectly valid.  

```python
c += 1
print('now c is: ', c)
```

Until now we have defined variables one by one each on its own line. But there is a more compact way to do this. In fact, we can combine any number of statements on a single line by separating them with semi-colons:  

```python
a=2; b=2; c=a+b
print(a,b,c)
```

And here is another way to do the exact same thing: 

```python
d, e, f = 4, 5, 6
print(d, e, f)
```
