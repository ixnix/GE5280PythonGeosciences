## List
https://docs.python.org/3/tutorial/datastructures.html

- Lists are denoted with **[ ]** and can contain any arbitrary set of elements, including other lists!
- Elements in the list can be indexed and sliced, similar to strings.
- The indices begin at 0. You can also count from the end to the beginning by starting with -1 (the last item in the list), -2 (second to last), etc. 
- Lists have **methods** that allow items to be sorted, deleted, inserted, sliced, counted, concatenated, replaced, added on to, etc.

The built-in methods of list:

| Method | Description |
|--------|---------|
|append()|Adds an element at the end of the list|
|clear()|	Removes all the elements from the list|
|copy()|	Returns a copy of the list|
|count()|	Returns the number of elements with the specified value|
|extend()|	Add the elements of a list (or any iterable), to the end of the current list|
|index()|	Returns the index of the first element with the specified value|
|insert()|	Adds an element at the specified position|
|pop()|	Removes the element at the specified position|
|remove()|	Removes the first item with the specified value|
|reverse()|	Reverses the order of the list|
|sort()|	Sorts the list|

Let's take a look at some examples: 

```python
mylist = [1,2,3,4] # defines a list
print(mylist) # prints the list

mylist = ['a','b','c','d'] # defines a list
print(mylist) # prints the list

mylist = ['a', 2.0, 2.0, '400', 'michigan', 42, [7, 8]] # defines a list
print(mylist) # prints the list
```

```python
print(mylist[2]) # print the third element in the list (starting from zero)
print(mylist[-1]) # print the last element
print (mylist[-3:]) # print the last three elements; NOTE THAT you cannot use mylist[-3:-1] because the slice list[begin:end] means from index = begin up to and not including index=end.
print (mylist[-3:-1])
```

Unlike strings, you can change list elements "in place". By "in place", we mean that you don't have to assign it to a new variable name; the list is changed "in place":

```python
mylist[3] = 'wisconsin'   # replaces the second element
print(mylist)
```

There are different ways to remove elements from a list:

- clear() method
- remove() method
- pop() method
- del statement

Let's look at a few examples,

```python
mylist = ['a', 2.0, 2.0, '400', 'michigan', 42, [7, 8]]
print(mylist)

mylist.clear() # remove all the elements
print(mylist)
```

```python
mylist = ['a', 2.0, 2.0, '400', 'michigan', 42, [7, 8]]
print(mylist)

mylist.remove(2.0) # remove the first element that is equal to 2.0
print(mylist)

mylist.remove('wisconsin') # this raises an error, because the specified value 'wisconsin' is not in the list.
print(mylist)
```

From the above, note that the _remove_ method only removes the **first** element for the specified value. If the specified value does not exist in the list, it's going to raise an error.

```python
mylist = ['a', 2.0, 2.0, '400', 'michigan', 42, [7, 8]]
print(mylist)

mylist.pop(3) # Remove the element at the specified index (3)
print(mylist)

mylist = ['a', 2.0, 2.0, '400', 'michigan', 42, [7, 8]]
elementRemoved = mylist.pop(3) # Note that the pop() method returns the element to be removed, not the resulted list.
print(elementRemoved)
```

```python
mylist = ['a', 2.0, 2.0, '400', 'michigan', 42, [7, 8]]
print(mylist)

del mylist[3] # You can also use the delete statement. Note the usage difference between pop() method and del statement 
print(mylist)
```

You can make a copy of the list by assigning it to another variable (like below), but this is known as a _shallow_ copy.

```python
mylist = ['a', 2.0, 2.0, '400', 'michigan', 42, [7, 8]]
mycopy = mylist 
print('copy: ', mycopy)

mylist[4] = 'wisconsin' # change the fifth element in the original list

print('original: ', mylist)
print('copy: ', mycopy)
```

Note that when you change the fourth element in the original list (mylist), the copy (mycopy) is also changed.
To spawn a list that is an independent object (a deep copy), you can do this: 

```python
mylist = ['a', 2.0, 2.0, '400', 'michigan', 42, [7, 8]]
mycopy = mylist[:]
print('copy: ', mycopy)

mylist[4] = 'wisconsin'

print('original: ', mylist)
print('copy: ', mycopy)
```

Now mycopy stayed the way it was, even if mylist changed.

To test if an element exists in the list, use the **in** keyword:

```python
mylist = ['a', 2.0, 2.0, '400', 'michigan', 42, [7, 8]]

'michigan' in mylist
```

```python
'illinois' in mylist
```

## Object

An object in Python is a collection that has _attributes_ and _methods_. The list variable _mylist_ is an example of an object - a _list_ object. In fact they are **classes** which we will learn more about in later lectures, but I just wanted to mention them here.

So what is so special about _objects_? Python objects have _methods_ which allow you to do things to the object. Methods have the form:

**object.method( )**

or 

**object.method(argument1, argument2,...)**

Here, **arguments** are something that can get passed into the method.

In the above, we have seen examples of list methods, e.g., clear(), pop(), remove(). Let's look at a few more handy methods of list.

- **.append()** method appends something (the **argument**) to the end of a list.
- **.count()** returns the number of times the argument occurs in the list.
- **.index( )** method returns the **index** of the desired argument in the list.

```python
mylist = ['a', 2.0, 2.0, '400', 'michigan', 42, [7, 8]]
print(mylist)

mylist.append('wisconsin') # add a new element 'wisconsin' to the end of the list
print(mylist)

print(mylist.count(2.0)) # count the number of times 2.0 appears in the list

print(mylist.index(2.0)) # print the index value of 2.0 in the list; Note that only the first index is returned, if there are multiple elements with the same value.
```

In the above, we made a list by defining a variable with square brackets. Another way to generate a list is to use **range( )**, which is one of the python _built-in functions_. The function **range(_start_,_end_,_step_)** is a _list generator_ and can be used to generate a list of integers between two numbers, the _start_ and _end_, where each number is separated by a _step_. You can make a lisa from the generator like so: **list(range(start,end,interval))**. Note that **range**, like **list** slicing, goes up to but does not include _end_. 

The **range** type represents an **immutable sequence** of numbers and is commonly used for looping a specific number of times in **for** loops.

To learn more about **immutable sequence types**, go [here](https://docs.python.org/3/library/stdtypes.html#typesseq).

Let's look at a few examples:

```python
print(list(range(10)))
print(list(range(3,10)))
print(list(range(0,10,2)))
```

## Tuple

https://docs.python.org/3/tutorial/datastructures.html#tuples-and-sequences

**Tuples** are another built-in structure in Python that are similar to lists, but have important differences. They are denoted by parentheses **( )** and consist of values separated by commas.

Like lists, they can contain different elements, but unlike lists, the elements cannot be changed in place. Their primary use is to pass information into and out of programs as we shall see in the coming lectures.  

Similar to both lists and strings, you can slice, concatenate, etc.

Here is one way to generate a tuple: 


```python
t = 1, 2.0, 'mtu'  # this is referred as 'tuple packing'
print(t)
```

```python
a,b,c = t # this is referred as 'tuple unpacking'
print(a)
print(b)
print(c)
```

You can access an element in a tuple by using the index number, exactly like a list.

```python
print(t[2])
```

But tuples are immutable - you cannot change it like this:

```python
t[2] = 'nmu'
```

## Set
https://docs.python.org/3/tutorial/datastructures.html#sets

Sets are denoted with curly braces **{ }**.

A set "contains an **unordered** collection of **unique** and **immutable** objects."   

You can create _sets_ in several ways. The first would be the use the python built-in **set( )** function on a list:

```python
mylist = ['mtu','nmu',10]
S1 = set(mylist)
print(S1)

# or just use the curly braces
S1 = {'mtu','nmu',10}
print(S1)
```

Notice how the order changed.

Also, notice what happens if we violate the "unique" part of the definition:


```python
S1 = set(['mtu','mtu','nmu',10])
print(S1)
```

Sets contain immutable objects, but they themselves can be changed. For a more complete list of methods see:  

https://www.python-course.eu/python3_sets_frozensets.php

But here are a few:

```python
S1 = {[1,2], ['a','b','c']} # Set doesn't allow mutable objects. Thus, we cannot include, for example, lists as elements
```

```python
# add
S1 = set(['mtu','mtu','nmu',10])
print(S1) 
S1.add('umich')

print(S1)
```

```python
# copy
S1 = set(['mtu','mtu','nmu',10])
S2 = S1.copy()
print('S2: ', S2)

S1.clear()
print('S1: ', S1)
print('S2: ', S2)
```

The **.copy( )** method for sets does not work like copying lists - it made an independent object **S2** which did not clear when **S1** got cleared.  

**.difference** is another handy method - it can be used to find what is different about two sets. 

```python
S1 = set(['mtu', 'nmu', 42])
S2 = set(['mtu', 'umich'])

# difference
print( S1.difference(S2) )

# or use a shortcut
print(S1-S2)
```

To create the **union** of the two sets,

```python
print(S1.union(S2))

# or use a shortcut
print(S1|S2)
```

If you want to know what is the **same** in two sets, use the **.intersection** method.  

```python
print( S1.intersection(S2) )

# or use a shortcut
print(S1&S2)
```

## Dictionaries

**dictionary** is denoted by **\{ \}**. They are somewhat like lists, but instead of integer **indices**, they use alphanumeric (letters and numbers) **keys**:

One way to define a **dictionary** is with the curly braces, keys and values syntax like this:

```python
# defines a dictionary of zip codes
zipcodes = {'Houghton':'12345','Hancock':'49930','Chassell':'49916'}
print(zipcodes)
```

The keys are 'Houghton', 'Hancock', and 'Chassell'. The values are '12345', '49930', and '49916'. Both the keys and values are of string type.

Python allows the values in a dictionary to be any type – string, integer, a list, another dictionary, boolean, etc. 

However, **keys must always be an immutable data types**, such as strings, numbers, or tuples. We will discuss immutable vs mutable objects later.

To return the value associated with a specific key, use square brackets with the key name:

```python
print(zipcodes['Houghton'])
```

 To change a key's value:

```python
zipcodes['Houghton'] = '49931'
print(zipcodes)
```

To add a new key and value:

```python
zipcodes['Calumet'] = '49913'
print(zipcodes)
```

You can also use the **update** method to change the value of existing keys or to add new (key,value) pairs. It can be used for merging two dictionaries. 
The usage is **dict.update(iterable)** - here the _iterable_ can be a dictionary or an iterable with key value pairs. You may wonder what is an **iterable** - feel free to search it up or wait till we discuss it in the future.

Let's look at an example,

```python
zipcodes = {'Houghton':'12345', 'Hancock':'49930', 'Chassell':'49916'}
print(zipcodes)

newzip = {'Houghton':'49931', 'Lake Linden': '49945'}

zipcodes.update(newzip)
print(zipcodes)
```

In the above example, the value associated with key 'Houghton' in _zipcodes_ is updated; Meanwhile, a new key:value pair from _newdata_ is added to _zipcodes_.

Like the other containers we learned, dictionaries also have **methods**. **dict.update()** is one of the dictionary's methods.
The three methods **dict.keys()**, **dict.values()** and **dict.items()** return view objects. They provide a dynamic view on the dictionary’s entries, which means that when the dictionary changes, the view reflects these changes.

```python
print(zipcodes.keys()) # returns the dictionary’s keys

print(zipcodes.values()) # returns the dictionary’s values

print(zipcodes.items()) # returns the dictionary’s items
```

You can also convert the output to lists via:

```python
print(list(zipcodes.keys())) # returns an unordered list of the keys

print(list(zipcodes.values())) # returns an unordered list of the values

print(list(zipcodes.items())) # returns an unordered list of the items as (key, value) tuples
```

How do we sort dictionaries? Does it have a **sort()** method similar to lists? Unfortunately, no.

```python
print(zipcodes.sort())
```

But, for a **sorted** list of keys, you can use the built-in function, **sorted**.  

```python
print(sorted(zipcodes.keys()))
```

To find the number of (key, value) pairs in the dictionary, use the **len** function:

```python
print(len(zipcodes))
```

There are other ways to create dictionaries. The **dict( )** constructor can make a dictionary directly from a **list** containing key-value **tuples**:  

```python
mylist = [('Michigan Technological University','MTU'), ('Michigan State University','MSU'), ('University of Michigan','UM')]
print(mylist)

dictName = dict(mylist)
print(dictName)

dictLocation = dict([('MTU','Houghton, MI'),('MSU','Lansing, MI'),('UM','Ann Arbor, MI')])
print(dictLocation)
```

To find the location of Michigan Technological University?

```python
print( dictLocation[ dictName['Michigan Technological University'] ] )
```

## Python data types and structures: Immutable vs. Mutable

Read complete article here: https://towardsdatascience.com/immutable-vs-mutable-data-types-in-python-e8a9a6fcfbdc

Everything in Python is an object. Objects are abstraction for data. 

A first fundamental distinction that Python makes on data is about whether or not the value of an object changes. If the value can change, the object is called mutable, while if the value cannot change, the object is called immutable.

To get started, it’s important to understand that every object in Python has an ID (or identity), a type, and a value, as shown below:

```python
age = 42
print(id(age))
print(type(age))
print(age)
```

Once created, the ID of an object never changes. It is a unique identifier for it, and it is used behind the scenes by Python to retrieve the object when we want to use it.

The type also never changes. The type tells what operations are supported by the object and the possible values that can be assigned to it.

The value can either change or not. If it can, the object is said to be **mutable**, while when it cannot, the object is said to be **immutable**.

What if we assign a new value to age, say, 43?

```python
age = 43
print(id(age))
print(type(age))
print(age)
```

Has the value of age changed? Well, no. 

42 is an integer number, of the type int, which is immutable. So, what happened is really that on the first line, age is a name that is set to point to an **int object**, whose value is 42.

When we type age = 43, what happens is that another object is created, of the type int and value 43 (also, the id will be different), and the name age is set to point to it. So, we didn’t change that 42 to 43. We actually just pointed age to a different location.

As you can see from printing id(age) before and after the second object named age was created, they are different!

Let's now look at an example of mutable object, e.g., a list.

```python
mylist = [1, 2, 3]
print(mylist)
print(id(mylist))

mylist.pop()
print(mylist)
print(id(mylist))
```

For this example, we created a list named _mylist_ that contains 3 integers. After we change _mylist_ by “popping” off the last element, the ID of _mylist_ stays the same!

So, objects of type int are immutable and objects of type list are mutable. 

Here is a summary of the immutable and mutable objects in Python:
- Mutable: **lists, byte arrays, sets, and dictionaries**
- Immutable: **numeric data types, strings, bytes, frozen sets, and tuples**
