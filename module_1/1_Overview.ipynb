Why do we need different environments? Conda environments are self-contained, isolated spaces where you can install specific versions of software packages, including dependencies, libraries, and Python versions. ## Google Colab (Colaboratory)

[Google Colab](https://colab.research.google.com/) is a free, browser-based [Jupyter notebook](https://jupyter.org/) environment hosted on Google's cloud. Students need only a Google account and a web browser — no local installation. Notebooks are stored in [Google Drive](https://drive.google.com/), can be shared like Google Docs, and run on Google-provided VMs with optional free [GPU](https://en.wikipedia.org/wiki/Graphics_processing_unit) and [TPU](https://cloud.google.com/tpu) access. Most major data science libraries — [NumPy](https://numpy.org/), [pandas](https://pandas.pydata.org/), [TensorFlow](https://www.tensorflow.org/), [PyTorch](https://pytorch.org/), [scikit-learn](https://scikit-learn.org/), [Matplotlib](https://matplotlib.org/) — come pre-installed. A free tier covers most coursework, with paid [Colab Pro and Pro+](https://colab.research.google.com/signup) tiers offering longer runtimes and better hardware allocation.

### How it differs from Anaconda/Miniconda

The fundamental difference is **where the code runs**: [Anaconda](https://www.anaconda.com/products/distribution) and [Miniconda](https://docs.conda.io/projects/miniconda/en/latest/) execute Python on the student's own machine, while Colab executes on a remote Google VM accessed through a browser. Local installs give persistent, user-controlled environments managed via [`conda`](https://docs.conda.io/); Colab provides ephemeral VMs that reset after inactivity, with extra packages reinstalled per session via `!pip install`. Local environments use whatever CPU and RAM the student has; Colab provides cloud GPUs and TPUs that most students wouldn't have access to locally. Files in Colab persist in Google Drive, but the runtime environment does not.

| Feature | Anaconda / Miniconda | Google Colab |
|---|---|---|
| Where code runs | Student's local machine | Google cloud VM |
| Installation needed | Yes (download + install) | None (browser only) |
| Environment persistence | Persistent, user-controlled | Ephemeral, resets per session |
| Package manager | `conda`, `pip` | `!pip install` per session |
| GPU/TPU access | Only if student owns hardware | Free tier (subject to limits) |
| Works offline | Yes | No |
| File persistence | Local filesystem | Google Drive |

### Advantages

Colab eliminates the setup barrier — students can be writing and running Python within minutes of clicking a link, with every student getting an identical environment. Free GPU and TPU access makes [deep learning](https://en.wikipedia.org/wiki/Deep_learning) exercises feasible without expensive hardware. Sharing and collaboration are trivial: instructors distribute assignments as links, and students can share notebooks for review. It works identically across Windows, Mac, Linux, [Chromebooks](https://www.google.com/chromebook/), and even tablets, lowering the equipment requirement for any course.

### Disadvantages

Colab requires reliable internet — offline work isn't possible. Sessions time out after periods of inactivity (typically **90 minutes idle, 12 hours maximum** on the free tier), losing in-memory state. Free-tier GPU access varies with demand and can be throttled. It's poorly suited for large datasets, long-running jobs, or projects needing system-level configuration. Code and data run on Google's infrastructure, which raises privacy considerations for sensitive research. It also doesn't teach students how to manage local Python environments — a skill they'll need for real-world projects beyond the classroom.

### Why it's useful for learning Python

Colab removes the single most common barrier to entry: getting Python installed and working before the real learning can begin. Students can focus on language fundamentals and concepts rather than troubleshooting environment issues in the first weeks of a course. Because there's no local state to break or reset, experimentation is frictionless. It pairs well with structured tutorials, since instructors can pre-build notebooks containing code, explanations, and exercises in a single document that students simply open and run.

> **Mental model for students:** Anaconda and Miniconda give you Python *on your computer*; Google Colab gives you Python *in your browser*, running on Google's machines. Use Colab to **learn quickly and experiment**; use a **local conda install** when you need persistence, control, or are working with sensitive data.

## Operating system commands

Macs and PCs both have functions that can be called from a _command line_, such as listing the contents of a folder or file, creating new folders, changing permissions on files or folders, combining the contents of files, moving files and folders around, and so on. These commands are directed to the operating system instead of the Python interpreter. To make these actions independent of your particular operating system, python has a built-in tool kit called "os" for operating system. We imported this in the first cell and will now figure out how to use this. 

Let's learn our first operating system command, which lists the contents of a directory, **os.listdir()**. This returns a list (not in any particular order) of all the things in the directory containing this notebook:

```python
os.listdir()
```

Another useful command is **os.mkdir()** which creates a new directory.

```python
os.mkdir('new_folder')
```

To see if that worked, list the contents again:

```python
os.listdir()
```

The command **os.rmdir()** deletes a directory 

```python
os.rmdir('new_folder')
```

Make sure it was removed:  

```python
os.listdir()
```

To delete a file (analogous to deleting a directory), we use the command **os.remove( )**.

## Concept of path

So far, we have looked at directories in our working directory (the one with this notebook in it) and subdirectories within the working directory. What if you need to access a file in any directories on your computer? You will have to first find the **path** for the file. You also need to know the difference between **absolute paths** and **relative paths**.

```python
Image(filename='img/filesystem.png',width=600) 
```

As an example, **/Users/xinxi-adm/Video/20210122_142801.jpg** is an example of _absolute path_, which uniquely define the location of the file or directory from anywhere on the computer.
The _relative paths_ are handy short cuts. For example, we can refer to a directory above the current directory without knowing what that is necessarily, we use these conventions:  
- ./ is the current directory
- ../ is the one above
- ../../ is the one above that 

and so on.

To find out what the _absolute path_ for your current directory, use **os.getcwd( )** to get the current working directory:

```python
os.getcwd()
```

To find the path to your home directory, we use another Python command, **Path.home()**. To use this, we already imported the toolkit **Path** in the first cell of this notebook so we are all set.

```python
Path.home()
```

And use that in the **os.listdir( )** command to get a listing of our home directory:

```python
os.listdir(Path.home())
```

To move change the name of a directory to another name, use the command **os.rename( )** (for rename). 

```python
os.mkdir('TEMP')
print('After creating a TEMP folder: ', os.listdir())

os.rename('TEMP','NEWTEMP')
print('After renaming the folder: ', os.listdir())

os.rmdir('NEWTEMP')
print('After removing the folder: ',os.listdir())
```
